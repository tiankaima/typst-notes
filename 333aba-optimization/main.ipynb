```matlab
clear all;

pList = [0.5,0.75,1,1.5,1.75,2];

for i=1:6
    subplot(2,3,i);
    p = pList(i);


    targetFunction = @(x) -0.5*x+1;
    x = -2:0.1:2;
    y = targetFunction(x);
    plot(x, y);
    hold on;

    measureFunction = @(x,y) abs(x).^p + abs(y).^p;
    C=min(measureFunction(x,y));
    fimplicit(@(x,y) measureFunction(x,y) - C, 'k');

    axis on;
    grid on;
    daspect([1 1 1]);
    axis([-1.5 1.5 -1.5 1.5]);
    title(['p = ', num2str(p)]);
    hold off;
end

exportgraphics(gcf, './1078ca-2023_fall_optimization/imgs/1.jpg', 'Resolution',500);
```
